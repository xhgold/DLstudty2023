```python
import tensorflow as tf
import numpy as np
tf.random.set_seed(777) # for reproducibility
```

```python
xy = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
# Check data
print(x_data,"\nx_data shape:", x_data.shape)
print(y_data, "\ny_data shape:", y_data.shape)
```

```python
X = tf.keras.Input(shape=(3, ), dtype=tf.float32)
Y = tf.keras.Input(shape=(1, ), dtype=tf.float32)

W = tf.Variable(tf.random.normal([3, 1]), name = 'weight')
b = tf.Variable(tf.random.normal([1]), name = 'bias')
```

```python
X = x_data
Y = y_data
```

```python
# hypothesis
def hypothesis(X):
    return tf.matmul(X, W) + b

# cost function
def cost_function(hypothesis, Y):
    return tf.reduce_mean(tf.square(hypothesis - Y))

# optimize/minimize
optimizer = tf.optimizers.SGD(learning_rate = 1e-5)
```

```python
# training
for step in range(2001):
    with tf.GradientTape() as tape:
        current_cost = cost_function(hypothesis(X), Y)
    gradients = tape.gradient(current_cost, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
    
    if step % 20 == 0:
        print(step, "Cost:", current_cost.numpy())
```
