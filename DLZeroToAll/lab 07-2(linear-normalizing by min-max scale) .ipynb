```python
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import MinMaxScaler
tf.random.set_seed(777)
```

```python
xy = np.array(
    [
        [828.659973, 833.450012, 908100, 828.349976, 831.659973],
        [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
        [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
        [816, 820.958984, 1008100, 815.48999, 819.23999],
        [819.359985, 823, 1188100, 818.469971, 818.97998],
        [819, 823, 1198100, 816, 820.450012],
        [811.700012, 815.25, 1098100, 809.780029, 813.669983],
        [809.51001, 816.659973, 1398100, 804.539978, 809.559998],
    ]
)
```

```python
scaler = MinMaxScaler()
xy = scaler.fit_transform(xy)
print(xy)
```

```python
x_data = xy[:, :-1]
y_data = xy[:, [-1]]
```

```python
X = tf.constant(x_data, dtype=tf.float32)
Y = tf.constant(y_data, dtype=tf.float32)

W = tf.Variable(tf.random.normal([4, 1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
# hypothesis
def hypothesis(X):
    return tf.matmul(X, W) + b

# cost_function
def cost_function(hypothesis, Y):
    return tf.reduce_mean(tf.square(hypothesis - Y))

# opmtimize
optimizer = tf.optimizers.SGD(learning_rate = 0.1)
```

```python
# trainiing
for step in range(2001):
    with tf.GradientTape() as tape:
        current_cost = cost_function(hypothesis(X), Y)
    gradients = tape.gradient(current_cost, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
    
    if step % 20 == 0:
        print(step, "Cost:", current_cost.numpy(), hypothesis(X).numpy())
```
